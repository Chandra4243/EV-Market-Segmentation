```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = df.rename(columns = {'Year (As on 31st March)' : 'Year', 'Buses @' : 'Buses', 'Two Wheelers*':'Bikes'})
```

```python
df.head()
print(df.columns)
```

```python
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 8)
```

```python
df.drop(columns=['All Vehicles'], inplace=True)
print("Successfully deleted the 'All Vehicles' column.")


plt.figure(figsize=(12, 8))

plt.plot(df['Year'], df['Bikes'], label='Bikes', marker='o', markersize=4)


plt.plot(df['Year'], df['Cars, Jeeps and Taxis'], label='Cars, Jeeps, Taxis', marker='o', markersize=4)

plt.plot(df['Year'], df['Goods Vehicles'], label='Goods Vehicles', marker='o', markersize=4)

plt.title('Growth of Key Vehicle Categories Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Registered Vehicles (in thousands)')
plt.legend()
plt.grid(True)
plt.show()

print("\nPlot created successfully!")
```

```python
# Calculate the correlation matrix
# We will drop 'Year' for this plot to focus on the vehicle relationships
correlation_matrix = df.drop('Year', axis=1).corr()

# Create the heatmap
plt.figure(figsize=(10, 7))
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt=".2f")
plt.title('Correlation Matrix of Vehicle Types')
plt.show()
```
