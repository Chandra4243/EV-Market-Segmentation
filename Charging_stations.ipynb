Charging Stations

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("RS_Session_258_AU_1241_3.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df_sorted = df.sort_values(by="Charging Stations", ascending=False)

plt.figure(figsize=(10,6))
bars = plt.bar(df_sorted['City/Highway'], df_sorted['Charging Stations'])

plt.title("Charging Stations by City/Highways")
plt.xlabel("City")
plt.ylabel("Charging Stations")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Charging stations by states

```python
data = pd.read_csv("RS_Session_267_AU_3266_2_0.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
data_sorted = data.sort_values(by="EV Charging stations", ascending=False)

plt.figure(figsize=(15,8))
bars = plt.bar(data_sorted["State/UT"], data_sorted['EV Charging stations'])

plt.title("Charging Stations by State/UT")
plt.xlabel("State/UT")
plt.ylabel("Charging Stations")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
