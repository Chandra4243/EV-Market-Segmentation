```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Ev_by_place.csv")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

top10 = df.sort_values("Market Share (%)", ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(data=top10, x="Market Share (%)", y="EV Maker", palette="viridis")
plt.title("Top 10 EV Makers by Market Share", fontsize=14)
plt.xlabel("Market Share (%)")
plt.ylabel("EV Maker")
plt.show()
```

```python
state_cs = df.groupby("State")["Charging Stations"].sum().sort_values(ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(x=state_cs.values, y=state_cs.index, palette="magma")
plt.title("Charging Stations by State", fontsize=14)
plt.xlabel("Number of Charging Stations")
plt.ylabel("State")
plt.show()
```

```python
plt.figure(figsize=(15, 15))
df.groupby("EV Maker")["Market Share (%)"].sum().plot(
    kind="pie", autopct="%1.1f%%", cmap="Set3"
)
plt.ylabel("")
plt.title("Market Share Distribution", fontsize=14)
plt.show()
```
