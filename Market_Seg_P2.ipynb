```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Health-fw-stat-2015_A11.csv")
```

```python
df = df.drop(columns=["Sl.No."])
```

```python
df.head()
```

```python
df.head()
```

```python
print(df.columns)
```

```python
india = df[df["States/Uts"] == "India"].iloc[0, 1:]  # skip "States/Uts"

india.plot(kind="bar", figsize=(12,6))
plt.title("Population by Age Group in India")
plt.xlabel("Age Group")
plt.ylabel("Population")
plt.show()
```

```python
state = "Uttar Pradesh"
row = df[df["States/Uts"] == state].iloc[0, 1:]   # all columns except 'States/Uts'

row.plot(kind="bar", figsize=(12,6))
plt.title(f"Population by Age Group in {state}")
plt.xlabel("Age Group")
plt.ylabel("Population")
plt.show()
```

```python
import matplotlib.pyplot as plt

state = "Karnataka"
row = df[df["States/Uts"] == state].iloc[0, 1:]   # all columns except 'States/Uts'

row.plot(kind="bar", figsize=(12,6))
plt.title(f"Population by Age Group in {state}")
plt.xlabel("Age Group")
plt.ylabel("Population")
plt.show()
```

```python
states = ["India", "Uttar Pradesh", "Kerala", "Tamil Nadu"]
subset = df[df["States/Uts"].isin(states)].set_index("States/Uts")

subset.T.plot(kind="bar", figsize=(14,7))
plt.title("Age Distribution: India vs Selected States")
plt.xlabel("Age Group")
plt.ylabel("Population")
plt.legend(title="Region")
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(14,8))
sns.heatmap(df.set_index("States/Uts"), cmap="viridis")
plt.title("Population Heatmap by Age Group (India & States/UTs)")
plt.show()
```
